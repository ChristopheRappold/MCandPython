# Monte Carlo techniques and numerical estimations




Before going to the Monte Carlo methods and numerical methods for error propagation. Some reminders on Statistics are necessary to understand the purposes of Monte Carlo methods/simulations. 

## Estimators

An estimator is a function of data (statistic) used for estimating an unknown parameter in a statistical model. If the _estimand_, the parameter to be estimated, is $\theta$, the estimator is symbolized by $\hat{\theta}$. As said, an estimator is associated a dataset or more generally to a random variable $X$: $\hat{\theta}(X)$. There are point estimator and interval estimator.  

- The estimators have several characteristics that are important to be verified:

  1. Bias:
     
     $Bias(\hat{\theta}) = \mathrm{E}[\hat{\theta}] - \theta$
      
  2. Consistency:
     
     The estimator is consistent if it converges in probability to the true value of the parameter.
     
  3. Efficiency:
    
    An estimator that has minimun variance.

  4. Robustness:
    
    The capacity of the estimator to not be modified by small variations of the data or of the model used.


- Examples of classical estimators:
  for independently observed data $x_i$:
      
  + sample mean:
      
      $$\bar{x} = \frac{1}{n} \sum x_i $$
      
  + sample variance:
          
      $$S^2 = \frac{1}{n-1} \sum (x_i - \bar{x})^2 $$
      
  + population variance:
  
      $$\sigma^2 = \frac{1}{n} \sum (x_i - \mu)^2 $$
      
      when the expectation of the population, $\mu$ is known
      

 
## Probability density function

A probability density function $f_X$ is a function that presents a probability distribution $X$ as an integral :  

$$\mathrm{Pr}[X\leq x]=\int_{-\infty}^x f_X(u)\mathrm{d}u$$

$$\mathrm{Pr}[a\leq X\leq b]=\int_{a}^b f_X(u)\mathrm{d}u$$

The cumulative density function, $F_X$, is then defined as :

$$ F_X(x) = \int_{-\infty}^x f_X(u)\mathrm{d}u$$

The function f(x) must be positive and normalized :

$$\int_{-\infty}^{+\infty} f_X(u)\mathrm{d}u = 1$$

### Important discrite probability distribution :

#### Bernoulli distribution:
    
When one observation drawn is a (success/fail) (yes/no) event. $p$ being the probability of obtaining "success/yes" and $1-p$ the probability of (fail/no)
    
$$ Pr [X=x] = p^x(1-p)^{1-x} \ with\ x = (1,0) $$

Its mean and variance are : $p$ and $p(1-p)$   

#### Binomial distribution:

When n observations of a Bernoulli law experiment with $p$ of probability associated to the success and the observation of $k$ success:
    
$$ Pr [X=k] = \binom{n}{k} p^k(1-p)^{1-k}$$

Its mean and variance are : $n\cdot p$ and $n\cdot p(1-p)$

![alt text](https://upload.wikimedia.org/wikipedia/commons/thumb/7/75/Binomial_distribution_pmf.svg/300px-Binomial_distribution_pmf.svg.png "Binomial distribution")

#### Poisson distribution:
When events occur independently with a certain rate, independently of the time, within a certain fixed interval of time

$$ Pr [k\ events] = \mathrm{e}^{-\lambda}\ \frac{\lambda^k}{k!}$$

where $\lambda$ is the average rate during this fixed interval
Its mean and variance are : $\lambda$ and $\lambda$   

![alt text](https://upload.wikimedia.org/wikipedia/commons/thumb/c/cb/Fonctions_de_masse_de_la_loi_de_Poisson.png/280px-Fonctions_de_masse_de_la_loi_de_Poisson.png)

#### Uniform distribution
    
When observations are considered to be uniformly distributed, example fair dice throw or n integer in interval [a,b] 


$$ Pr = 1/n$$

with $n = b -a+1$

Its mean and variance are : $\frac{a+b}{2}$ and $\frac{(b-a+1)^2-1}{12}$

![alt text](https://upload.wikimedia.org/wikipedia/commons/thumb/1/1f/Uniform_discrete_pmf_svg.svg/325px-Uniform_discrete_pmf_svg.svg.png)



### Important continous probability distribution :

#### Uniform distribution

When observations are considered to be uniformly distributed in the interval [a,b]

$ f(x) = 1/(b-a) $ for $a\leq x\leq b$ and $f(x) = 0$ otherwise

Its mean and variance are : $\frac{a+b}{2}$ and $\frac{(a-b)^2}{12}$

![alt text](https://upload.wikimedia.org/wikipedia/commons/thumb/9/96/Uniform_Distribution_PDF_SVG.svg/250px-Uniform_Distribution_PDF_SVG.svg.png)

#### Exponential distribution 

When observation of the time between Poisson law based observations occur

$f(x, \lambda) = \lambda\ \mathrm{e}^{-\lambda}$ for $x\geq 0$

where $\lambda$ is the rate parameter

alternative expression:

$f(x, \tau) = 1/\tau\ \mathrm{e}^{-1/\tau}$ for $x\geq 0$

where $\tau$ is the survivial parameter, or decay time

Its mean and variance are : $1/\lambda$ or $\tau$ and $1/\lambda^2$ or $\tau^2$

![alt text](https://upload.wikimedia.org/wikipedia/commons/thumb/e/ec/Exponential_pdf.svg/325px-Exponential_pdf.svg.png)

#### Gaussian distribution

The most probability distribution associated to physical occurence, either intrinsically or approximatelly via the central limit theorem.

$$ f(x\ |\ \mu, \sigma^2) = \frac{1}{\sqrt{2\pi\sigma^2}}\ \exp{(-\frac{(x-\mu)^2}{2\sigma^2})}$$

Its mean and variance are : $\mu$ and $\sigma^2$

![alt text](https://upload.wikimedia.org/wikipedia/commons/thumb/7/74/Normal_Distribution_PDF.svg/350px-Normal_Distribution_PDF.svg.png)


##### Central limit theorem

Supposing a set of $n$ random variable $X_i$ n independent and identically distributed with mean, $\mu$, and variance $\sigma^2$:
$$S_n = \sum X_i $$
$$\bar{X_n} = S_n/n$$
$$Z_n = \frac{\overline{X_n} - \mu}{\sigma/\sqrt{n}}$$

For $n$ large :
$$\overline{X_n} \rightarrow \mathrm{N}(\mu, \sigma^2/n),\ S_n \rightarrow \mathrm{N}(n\mu, n\sigma^2),\ Z_n \rightarrow N(0,1)$$


![alt text](https://image.jimcdn.com/app/cms/image/transf/dimension=571x10000:format=png/path/saae1248de6782df1/image/i9ab857a3c7ecec9c/version/1476391200/image.png)

+ Examples:
  - Sum of Bernoulli(p) $\rightarrow$ Binomial(n,p)
  - When $np$ fixed and $n\rightarrow\infty$ Binomial(n,p) $\rightarrow$ Poisson($\lambda$) with $\lambda=np$
  - When $n\rightarrow\infty$ with $\lambda=\sigma^2$ Poisson($\lambda$) $\rightarrow$ Normal($\mu$,$\sigma^2$)
  - When $n\rightarrow\infty$ with $\mu=np$ and $\sigma^2=np(1-p)$ Binomial(n,p) $\rightarrow$ Normal($\mu$,$\sigma^2$)
  - Sum of Uniform(a,b) $\rightarrow$ N($\mu$,$\sigma^2$)

## Monte Carlo methods 

Monte Carlo methods are numerical algorithm that are based on random sampling to obtain a numerical estimations or results. 

The expected value of function $g$ of a random variable X that follows a pdf $f_X$ in the interval [a,b]:

$$G = E(g(X)) = \int g(x)f_X(x)\mathrm{d}x$$

let say that $f_X$ is a uniform distribution: $f_X(x) = 1/(b-a)$ 
If you sample $(x_i)$ from X, then you can estimate G.
By defining the following estimator:
$$ \tilde{g}(n) = \frac{1}{n} \sum g(x_i) $$

And with the central limit theorem:

$$ \frac{\tilde{g}(n) - G}{\sigma_g / \sqrt{n}} \rightarrow N(0,1)$$

where $\sigma_g^2 = S^2_{g(X)} = \int g^2(x)f_X(x)dx - G^2$

We see that the error of estimation is of the order of $1/\sqrt{n}$, $n$ being the numbers of sample of the distribution of random variable X.  


### Integration of difficult high-dimensions integrals

The classic numerical methods for integration have a real diffculties to estimates integral of high dimensions:
+ in 1-D: 10 evaluations of Simpson's rules : errors $\sim 10^{-7}$ 
+ in 10-D: needs $10^{10}$ evaluations for similar errors 

With Monte Carlo method such integral can be calculated.

![alt text](http://web-docs.gsi.de/~crappold/rev-monte-carlo_Int2.png)

$$I = \int_\Omega f(\mathbf{x}) d\mathbf{x}$$

1. In 1-D: Sample random uniformly distributed ($x_i$, $y_i$) sets within the dashed box, if $y_i<f(x_i)$, the sample is accepted, otherwise it is rejected. Then the integral is $I=\frac{accepted}{rejected}$   

2. In k-D: $I$ can be approximated by : $I_n = V\ \frac{1}{N} \sum f(x_i) = V\ E(f)$ with $V = \int_\Omega d\mathbf{x}$ the volume of of $\Omega$.
 Thanks to the central limit theorem We have: $I_n \rightarrow I$ with an error decreasing of the order of $1/\sqrt{N}$

Example:



```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.integrate import quad

%matplotlib inline

def circle(x, y):
    return np.sqrt(x**2+y**2)<=1

def circle2(x):
    return np.sqrt(1-x**2)

xgrid = np.linspace(-1, 1, 500)
ygrid = np.linspace(-1, 1, 500)
x, y = np.meshgrid(xgrid, ygrid)

xgrid2 = np.linspace(0, 1, 500)
z2 = circle2(xgrid2)

z = circle(x,y)
max_y = 1.2
x_rand = np.random.uniform(0,1,1000000)
y_rand = np.random.uniform(0,max_y,1000000)

x_acc = x_rand[y_rand<=circle2(x_rand)]
x_rej = x_rand[y_rand>circle2(x_rand)]

y_acc = y_rand[y_rand<=circle2(x_rand)]
y_rej = y_rand[y_rand>circle2(x_rand)]

sampling_plot_acc = np.random.randint(0,np.size(x_acc),1000)
sampling_plot_rej = np.random.randint(0,np.size(x_rej),1000)

fig7, ax7 = plt.subplots(figsize=(8,8))
ax7.plot(xgrid2,z2,linewidth=5)
ax7.scatter(x_acc[sampling_plot_acc],y_acc[sampling_plot_acc],c='r')
ax7.scatter(x_rej[sampling_plot_rej],y_rej[sampling_plot_rej],c='g')
plt.show()

est_pi = 4*np.size(x_acc)/np.size(x_rand) * max_y
print("pi:",np.pi)
print("est_pi = ",est_pi)

int2 = np.mean(circle2(x_rand))
int3, int3_err = quad(circle2,0,1)
print("est_pi2 =", 4*int2)
print("numerical = ",4*int3)

x_rand2 = np.random.uniform(-1.2,1.2,100000)
y_rand2 = np.random.uniform(-1.2,1.2,100000)

z = circle(x_rand2,y_rand2)

xr3 = x_rand2[z>0]
yr3 = y_rand2[z>0]
xr4 = x_rand2[z==False]
yr4 = y_rand2[z==False]

sampling_plot_acc2 = np.random.randint(0,np.size(xr3),1000)
sampling_plot_rej2 = np.random.randint(0,np.size(xr4),1000)


fig7_1, ax7_1 = plt.subplots(figsize=(8,8))
ax7_1.scatter(xr3[sampling_plot_acc2],yr3[sampling_plot_acc2],c='r')
ax7_1.scatter(xr4[sampling_plot_rej2],yr4[sampling_plot_rej2],c='g')
plt.show()

```

### Random sampling of complex distributions

1. The transformation method, It allows to draw random number from any distribution:

     Given a cdf of distribution, $F$,$F(y)= \int_{-\infty}{y}f(x)dx$. We have $0\leq F(y) \leq 1$ so by drawing uniformly over F(y). Therefore $F^{-1}(x)$ will follow the $f$ distribution.
 
   + Example: 

     for exponential distribution: $p(y)=\lambda\exp{(-\lambda y)}$.
     
     $$F(y) = 1-\exp{(-\lambda y)} = x$$
     with $x \sim U(0,1)$
     
     $y(x) = -1/\lambda \ln(1-x)$ which is equivalent to $y(x) = -1/\lambda \ln(x)$ because  $x \sim U(0,1)$ and $1-x$ also is  $\sim U(0,1)$
      



```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats as stat
import scipy.optimize as opt

%matplotlib inline

def rev_exp(x, l):
    return -1/l*np.log(x)

def exp1(x,l,a):
    return a*l*np.exp(-l*x)

x_rand = np.random.uniform(0,1,10000)
y_rand = rev_exp(x_rand,0.1)
print("Generated exponential distribution from inverse method:")
fig8, ax8 = plt.subplots(figsize=(8,6))
n, bins, patch = ax8.hist(y_rand,bins=100)
plt.show()

#print(n,"\n", bins)
bins_width = bins[1]-bins[0]
popt, pcov = opt.curve_fit(exp1, bins[:-1]+bins_width*0.5, n, p0= [0.3,300])
print("fitted value:",popt,"\nCov:\n",pcov)
#print(np.shape(n)," ",np.shape(bins))
print("\nfitted PDF with generated exponential distribution:")
fig8_1, ax8_1 = plt.subplots(figsize=(8,6))
ax8_1.hist(y_rand,bins=100)
ax8_1.plot(bins[:-1]+bins_width*0.5, exp1(bins[:-1]+bins_width*0.5, popt[0], popt[1]), 'r-', label='fit')
plt.show()


```

2. Acceptance-rejection method (Von Neumann)

 ![alt text](http://web-docs.gsi.de/~crappold/rev-monte-carlo_Int2.png)

 If the $F^{-1}(x)$ is not accessible because $F$ is unknown or too complex the first method is not possible. 
 
   + A known distribution $h(x)$ must be defined as such that the $f(x)$ is inside the $C$ times the $h(x)$ distribution: $C\ h(x) \geq f(x)$.
   + Generate a value $x$ following $h(x)$ and calculated $f(x)$ and $C\ h(x)$
   + Generate a second value $u$ within $U(0,1)$ and test $u\ C\ h(x) \leq f(x)$
   + If true then accept $x$ otherwise reject $x$ and retry.
   
 Those points $(x,u)$ populate the area $C\ h(x)$ and everything under $f(x)$ is accepted.

## Error propagation

How to propagate uncertainties on observables through functions and other combinaisons of those observable ? and if the observables are correlated ?

1. Linearisation around the mean value:
  
   +  you have X: $\mu_X$ and $\sigma_X^2$, with Taylor expansions 1-D:
      $$E[\ f(X)\ ] \approx f(\mu_X) + \frac{f''(\mu_X)}{2}\sigma_X^2$$
      and
      $$Var[\ f(X)\ ] \approx  (f'(\mu_X))^2\sigma^2$$
   + you have X $\mathbf{\mu_X}$ and covariance matrix $\Sigma(X)$ With Taylor expansions n-D:
      $$E[\ f(X)\ ] \approx f(\mu_X) + \frac{ 1}{2!} E(X-\mu_X)^T D^2f(\mu_X) E(X-\mu_X)$$
      and
      
      $$Var[\ f(X)\ ] \approx Df(\mu_X)\cdot\Sigma_X\cdot Df(\mu_X)^T$$
      or
      $$Var[\ f(X)\ ] \approx J\cdot\Sigma_X\cdot J^T $$  
      
    $$\Sigma_X = \begin{pmatrix}
    \sigma_1^2 & \sigma_{12} & \sigma_{13} & \cdots \\ 
    \sigma_{12} & \sigma_2^2 & \sigma_{23} & \cdots\\ 
    \sigma_{13} & \sigma_{23} & \sigma_3^2 & \cdots\\ 
    \vdots & \vdots & \vdots & \ddots 
    \end{pmatrix} $$

    Example: $f(z)= X/Y$

    The jacobian is: $J = (1/Y,\ -X/Y^2)$, $\Sigma_{XY} = \begin{pmatrix} \sigma_x^2 & \sigma_{xy} \\ \sigma_{xy} & \sigma_y^2 \end{pmatrix}$

    Therefore : 
    $$\sigma_f^2 \approx J\cdot \Sigma_{XY} \cdot J^T$$
    $$\sigma_f^2 \approx \frac{ \sigma_x^2}{y^2} -2\frac{x}{y^3} \sigma_{xy} + \frac{x^2}{y^4}\sigma_y^2$$
    $$\sigma_f^2 \approx f^2\cdot(\frac{\sigma_x^2}{x^2} + \frac{\sigma_y^2}{y^2} -2\frac{\sigma_{xy}}{xy})$$

2. $\ln$ Tricks :
  + Take your expression and apply $ln$ to it.
  + Do the differentation of the expression
  + Get an expression : df = K()
  + Take the square of it : df^2 = (K())^2
  + Associate df^2 to $\sigma_f^2$, dx^2 to $\sigma_x^2$ and dxdy to $\sigma_{xy}$
  
  Example : $f(z) = X/Y$
    1. $ln(f) = ln(X) - ln(Y)$
    2. $\frac{df}{f} = \frac{dX}{X} - \frac{dY}{Y}$
    3. $df = f\cdot(\frac{dX}{X} - \frac{dY}{Y})$
    4. $df^2 = f^2\cdot(\frac{dX}{X} - \frac{dY}{Y})^2$
    5. $df^2 = f^2\cdot(\frac{dX^2}{X^2} + \frac{dY^2}{Y^2})^2 -2\frac{dXdY}{XY})$
    6. $\sigma_f^2 \approx f^2\cdot(\frac{\sigma_x^2}{x^2} + \frac{\sigma_y^2}{y^2} -2\frac{\sigma_{xy}}{xy})$
    
3. Monte Carlo method:

 + Generate the random sample of each estimator of the observables.
 + Apply your transformation correcting to the function.
 + Investigate the obtained distribution
 + Mean and Variance can be estimated from the obtained distribution.
 

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats as stat
import scipy.optimize as opt

%matplotlib inline

class RatioPDF:
    def __init__(self, mu_a, sig_a, mu_b, sig_b, corr_ab):
        self.mu_a = mu_a
        self.mu_b = mu_b
        self.sig_a = sig_a
        self.sig_b = sig_b
        #self.InvSa2 = 1./sig_a**2
        #self.InvSb2 = 1./sig_b**2
        self.corr_ab = corr_ab
        
        self.c = self.mu_a**2/self.sig_a**2 - 2*self.corr_ab*self.mu_a*self.mu_b/self.sig_a/self.sig_b + self.mu_b**2/self.sig_b**2
        
    def a(self, z):    
        return np.sqrt(z**2/self.sig_a**2 - 2*self.corr_ab*z/self.sig_a/self.sig_b + 1/self.sig_b**2)
    def b(self, z):
        return self.mu_a*z/self.sig_a**2 - self.corr_ab*(self.mu_a+self.mu_b*z)/self.sig_a/self.sig_b + self.mu_b/self.sig_b**2
    def d(self, z):
        return np.exp(((self.b(z))**2-self.c*(self.a(z))**2)/(2*(1-self.corr_ab**2)*(self.a(z))**2))
    
    def __call__(self, z,K):#, mu_a, sig_a, mu_b, sig_b, corr_ab):
        temp_a = self.a(z)
        temp_b = self.b(z)
        temp_d = self.d(z)
        temp1 = temp_b*temp_d/(np.sqrt(2*np.pi)*self.sig_a*self.sig_b*(temp_a)**3)
        new_z = temp_b/(np.sqrt(1-self.corr_ab**2)*temp_a)
        temp1 *= stat.norm.cdf(new_z)-stat.norm.cdf(-new_z)
        temp1 += np.sqrt(1-self.corr_ab**2)/(np.pi*self.sig_a*self.sig_b*temp_a**2)*np.exp(-self.c/(2*(1-self.corr_ab**2)))
        return K*temp1
    
    
def ratioPDFext(x, ratio, K):
    #ratio = RatioPDF(mu_a, sig_a, mu_b, sig_b, corr_ab)
    return K*ratio(x)

    
mean_xy = np.array([20,3])
cov_xy = np.array([[0.3, 0.24],[0.24, 0.2]])
        
    
ratio = RatioPDF(mean_xy[0],np.sqrt(cov_xy[0][0]),mean_xy[1],np.sqrt(cov_xy[1][1]),cov_xy[0][1]/np.sqrt(cov_xy[0][0])/np.sqrt(cov_xy[1][1]))

x_1 = np.linspace(0, 20, 500)
z_1 = ratio(x_1,1)#,mean_xy[0],np.sqrt(cov_xy[0][0]),mean_xy[1],
          #np.sqrt(cov_xy[1][1]),cov_xy[0][1]/np.sqrt(cov_xy[0][0])/np.sqrt(cov_xy[1][1]),1)


    

rv = stat.multivariate_normal.rvs(mean=mean_xy, cov=cov_xy, size= 100000)
print("X vs Y:")
fig10, ax10 = plt.subplots(figsize=(8,6))
ax10.hist2d(rv[:,0],rv[:,1],bins=100,cmap='RdBu')
#plt.colorbar()
#print(type(ax10))
ax10.set_xlabel("x")
ax10.set_ylabel("y")
plt.show()
print("Ratio X/Y via Monte Carlo:")
Z = rv[:,0]/rv[:,1]
fig11, ax11 = plt.subplots(figsize=(8,6))
n, bins, patch = ax11.hist(Z,bins=100)
#ax11[0][1].plot(x,z,'-b')
ax11.set_xlabel("Z")
ax11.set_ylabel("Counts")
plt.show()

bins_width = bins[1]-bins[0]
popt, pcov = opt.curve_fit(ratio, bins[:-1]+bins_width*0.5, n)

print("PDF of Ratio X/Y:")

fig12, ax12 = plt.subplots(figsize=(8,6))
ax12.plot(x_1,z_1,'-b')
#ax12.hist(Z,bins=100)
#ax12.plot(bins[:-1], ratio(bins[:-1], *popt), 'r-', label='fit')
plt.show()

print("fitted PDF of Ratio X/Y + Ratio Z=X/Y:")

fig12_1, ax12_1 = plt.subplots(figsize=(8,6))
#ax12.plot(x_1,z_1,'-b')
ax12_1.hist(Z,bins=100)
ax12_1.plot(bins[:-1]+bins_width*0.5, ratio(bins[:-1]+bins_width*0.5, *popt), 'r-', label='fit')
ax12_1.set_xlabel("Z")
ax12_1.set_ylabel("Counts")
plt.show()

real_mean = np.sum(x_1*z_1)/np.sum(z_1)
real_variance = np.sum((x_1-real_mean)**2*z_1)/(np.sum(z_1)-1)
#print("value of the fit:",*popt)
print("real value: (mean,var)=",real_mean, real_variance)
print("Monte Carlo Z: (mean,var)=",Z.mean(), Z.var())
#def varRatio(mean1,mean2,sig1,sig2,cov12):
#    return mean1**2/mean2**2*(sig1/mean1**2+sig2/mean2**2-2*cov12/mean1/mean2)


#print("Linearize Z: (mean,var)=", mean_xy[0]/mean_xy[1], varRatio(mean_xy[0],mean_xy[0],cov_xy[0][0],cov_xy[1][1],cov_xy[0][1]) )
print("Linearize Z: (mean,var)=", mean_xy[0]/mean_xy[1], mean_xy[0]**2/mean_xy[1]**2 * (cov_xy[0][0]/mean_xy[0]**2 + cov_xy[1][1]/mean_xy[1]**2 -2*cov_xy[1][0]/(mean_xy[0]*mean_xy[1])) )

x_2 = np.linspace(0, 20, 500)
z_21 = ratio(x_2,1)
z_22 = stat.norm.pdf(x_2,mean_xy[0]/mean_xy[1],mean_xy[0]**2/mean_xy[1]**2 * (cov_xy[0][0]/mean_xy[0]**2 + cov_xy[1][1]/mean_xy[1]**2 -2*cov_xy[1][0]/(mean_xy[0]*mean_xy[1])))
z_23 = stat.norm.pdf(x_2,Z.mean(), Z.var())
                   
fig12_2, ax12_2 = plt.subplots(figsize=(8,6))
ax12_2.plot(x_2,z_22,"r:",label="Linear")
ax12_2.plot(x_2,z_23,"g-.",label="MC")
ax12_2.plot(x_2,z_21,"b-",label="RealPDF")
ax12_2.set_xlim(2,12)
ax12_2.set_xlabel("x")
ax12_2.legend()
plt.show()
```
